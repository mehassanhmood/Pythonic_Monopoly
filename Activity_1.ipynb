#### Activity 1.6 : Auto Correlation :

```python
weather = pd.read_csv('C:/Users/Zahra/Downloads/weather.csv',parse_dates = True,index_col = 'Date',infer_datetime_format =True)
```

```python
weather['lag temperature'] = weather.Temperature.shift()
```

```python
x = weather.Temperature
```

```python
y = weather['lag temperature']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
weather.Temperature.autocorr(lag=1)
```

```python
weather.Temperature.autocorr(lag=2)
```

```python
weather.Temperature.autocorr(lag=3)
```

```python
# temperature correlation with x hour today and the same hour yesterday
weather.Temperature.autocorr(lag=24)
```

```python
# To understand Cycles.
import statsmodels as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# plot_acf() measures autocorrelation at different lags
# Highest autocorrelation at lag of 24 hours
plot_acf(weather.Temperature, lags=48);
```
